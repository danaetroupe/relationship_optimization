```python
# Data Processing
import pandas as pd
import numpy as np
import pyreadstat

# Modelling
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, ConfusionMatrixDisplay
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from scipy.stats import randint

# Tree Visualisation
from sklearn.tree import export_graphviz
from IPython.display import Image
import graphviz
```

```python
df, metadata = pyreadstat.pyreadstat.read_sav("S1Dataset.sav")
```

```python
# prompt: drop all columns where the row 'sat_sex_A' has a nan value

df = df.dropna(subset=['sat_sex_A'], axis=0)
```

```python
# Visualize unique values
for col in df.columns:
  print(f"Column: {col}")
  print(f"Unique values: {df[col].unique()}")
  print("-" * 20)
```

```python
# Split the data into features (X) and target (y)
X = df.drop(['sat_sex_A', 'sat_sex_P', 'Dyad_ID', 'partnum'], axis=1)
y = df['sat_sex_A']
# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
