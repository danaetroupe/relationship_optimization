```python
# this is to format string values in relshare-floats into float values
import pandas as pd
```

```python
def convert_column(file_name, column):
    df = pd.read_csv(file_name)

    if column not in df.columns:
        print(f'Errror: Column {column} does not exist.')
        return None

    unique_values = df[column].unique()
    value_to_int_map = {value: idx for idx, value in enumerate(unique_values)}
    df[column] = df[column].map(value_to_int_map)

    print("Mapping of unique values to integers:")
    for value, idx in value_to_int_map.items():
        print(f"{value} -> {idx}")

    return df
        
```

```python
file = "data-floats.csv"
#convert_column(file, "Age")
convert_column(file, "gender")
```
